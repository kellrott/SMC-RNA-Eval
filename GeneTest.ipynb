```python
import re
import os
import csv
import json
import random
import pandas
import itertools
from glob import glob
```

```python
info_file = "gene_transcript_exon.features"
```

```python
transcript_gene = {}
gene_transcript = {}
with open(info_file) as handle:
    reader = csv.DictReader(handle, delimiter="\t")
    for row in reader:
        transcript_gene[row["Transcript stable ID"]] = row["Gene stable ID"]
        s = gene_transcript.get(row["Gene stable ID"], set())
        s.add(row["Transcript stable ID"])
        gene_transcript[row["Gene stable ID"]] = s
```

```python
exp = {}
for i in glob("real_output_redo_strand/isoform/"):
    res = re.search("(\d+)_([\w_]+)_expression.tsv", i)
    if res:
        sub, cel = res.groups()
        if cel not in exp:
            exp[cel] = {}
        
```

```python
exp = {}

for i in glob("real_output_redo_strand/isoform/*/*/job.out"):
    print "Loading", i
    cel = os.path.basename(os.path.dirname(i))
    sub = os.path.basename(os.path.dirname(os.path.dirname(i)))
    with open(i) as handle:
        d = json.loads(handle.read())
        f = os.path.join(os.path.dirname(i), d['OUTPUT']['basename'])
        if cel not in exp:
            exp[cel] = {}
       
        df = {}
        with open(f, "rt") as a:
            for line in a:
                row = line.rstrip().split("\t")
                if row[0].startswith("ENST0"):
                    df[row[0]] = float(row[1])
        exp[cel][sub] = df
```

```python
dfs = {}
for k, v in exp.items():
    dfs[k] = pandas.DataFrame(v)
```
