```python
import pandas
```

```python
adminTeam = {"jeltje" : "SMC-Admin", "kellrott" : "SMC-Admin"}
```

```python
spike_leaderboard = pandas.read_csv("fusion_leaderboard.tsv", sep="\t", index_col=0)
```

```python
spike_gene_leaderboard = pandas.read_csv("fusion_gene_leaderboard.tsv", sep="\t", index_col=0)
```

```python
sim_leaderboard = pandas.read_csv("fusion_sim_leaderboard.tsv", sep="\t", index_col=0)
```

```python
entries = pandas.read_csv("results/Final_fusion_entries.tsv", sep="\t", index_col=0)
```

```python
notations = pandas.read_csv("results/Fusion_Methods.tsv", sep="\t", index_col=0)
```

```python
spike_leaderboard.join(sim_leaderboard)
```

```python
leaderboard = spike_leaderboard.join(sim_leaderboard).loc[ entries.index.intersection(spike_leaderboard.index) ]
leaderboard = leaderboard.join(notations[ ["Team", "EntryName", "Aligner", "Methods", "Github"] ])
leaderboard = leaderboard.join(spike_gene_leaderboard)
leaderboard.sort_values(by="exon_ave_f1", ascending=False)
```

```python
leaderboard.Team = leaderboard.Team.replace(adminTeam)
```

```python
leaderboard.to_csv("fusion_combined_leaderboard.tsv", sep="\t")
```
